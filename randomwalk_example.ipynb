```python
import numpy 
import matplotlib.pyplot as plt
import random 

# write the function that performs 
#a random walk with p players and n steps
def RW(p, n):
    x = numpy.zeros((p,n)) 
    y = numpy.zeros((p,n))
    for j in range(0,p):
        for i in range(1, n): 
            val = random.randint(1, 4) 
            if val == 1: 
                x[j][i] = x[j][i - 1] + 1
                y[j][i] = y[j][i - 1] 
            elif val == 2: 
                x[j][i] = x[j][i - 1] - 1
                y[j][i] = y[j][i - 1] 
            elif val == 3: 
                x[j][i] = x[j][i - 1] 
                y[j][i] = y[j][i - 1] + 1
            else: 
                x[j][i] = x[j][i - 1] 
                y[j][i] = y[j][i - 1] - 1
    return x,y
      
# Two example plots
fig = plt.figure()
plt.rcParams['figure.figsize'] = (10,8)
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)

n = 100
p = 100
sx,sy=RW(p,n)  
ax1.set_title("Random Walk with "+str(p)+" players ($n = " + str(n) + "$ steps)") 
for j in range(0,p):
    ax1.plot(sx[j], sy[j])
ax1.plot(0, 0, 'go')


n = 1000
p = 1000
sx,sy=RW(p,n)  
ax2.set_title("Random Walk with "+str(p)+" players ($n = " + str(n) + "$ steps)") 

for j in range(0,p):
    ax2.plot(sx[j], sy[j])
ax2.plot(0, 0, 'go')

```

```python
import numpy 
import matplotlib.pyplot as plt
import random 

# write the function that performs 
#a random walk with p players and n steps
#and a step size dx
def RW_dx(p, n, dx):
    x = numpy.zeros((p,n)) 
    y = numpy.zeros((p,n))
    for j in range(0,p):
        for i in range(1, n): 
            val = random.randint(1, 4) 
            if val == 1: 
                x[j][i] = x[j][i - 1] + dx
                y[j][i] = y[j][i - 1] 
            elif val == 2: 
                x[j][i] = x[j][i - 1] - dx
                y[j][i] = y[j][i - 1] 
            elif val == 3: 
                x[j][i] = x[j][i - 1] 
                y[j][i] = y[j][i - 1] + dx
            else: 
                x[j][i] = x[j][i - 1] 
                y[j][i] = y[j][i - 1] - dx
    return x,y
      
# Two example plots
fig = plt.figure()
plt.rcParams['figure.figsize'] = (10,8)
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)

n  = 100
p  = 100
dx = 0.1
sx,sy=RW_dx(p,n,dx)  
ax1.set_title("Random Walk with "+str(p)+" players ($n = " + str(n) + "$ steps)") 
for j in range(0,p):
    ax1.plot(sx[j], sy[j])
ax1.plot(0, 0, 'go')


n  = 1000
p  = 1000
dx = 0.01
sx,sy=RW_dx(p,n,dx)  
ax2.set_title("Random Walk with "+str(p)+" players ($n = " + str(n) + "$ steps)") 

for j in range(0,p):
    ax2.plot(sx[j], sy[j])
ax2.plot(0, 0, 'go')
```

# Ornstein–Uhlenbeck process
We are going to simulate a AR-1 process

\begin{equation} x_{n+1} = -K\,x_n + \varepsilon_n \end{equation}

where $K>0$ and $\varepsilon_n$ is the white noise (random walk). There is a potential on the walker that makes it moves back towards a central location, with a greater attraction when the walker is further away from the center. Higher $K$ means higher attraction to the center.

```python
import numpy as np
import matplotlib.pyplot as plt
import random 


def RW_OU(p, n, K):
    
    x = np.zeros((p,n)) 
    y = np.zeros((p,n))
    
    for j in range(0,p):
        for i in range(1, n): 
            val = random.randint(1, 4) 
            if val == 1: 
                x[j][i] = -K*x[j][i - 1] + 1 +x[j][i - 1]
                y[j][i] = y[j][i - 1] 
            elif val == 2: 
                x[j][i] = -K*x[j][i - 1] - 1 +x[j][i - 1]
                y[j][i] = y[j][i - 1] 
            elif val == 3: 
                x[j][i] = x[j][i - 1] 
                y[j][i] = -K*y[j][i - 1] + 1 +y[j][i - 1]
            else: 
                x[j][i] = x[j][i - 1] 
                y[j][i] = -K*y[j][i - 1] - 1 +y[j][i - 1]
    return x,y



# Two example plots
fig = plt.figure()
plt.rcParams['figure.figsize'] = (12,4)
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

n  = 500
p  = 2
K = 0
sx,sy=RW_OU(p,n,K)  
ax1.set_title("K="+str(K))

for j in range(0,p):
    ax1.plot(sx[j], sy[j])
ax1.plot(0, 0, 'go')


n  = 500
p  = 2
K = 0.15
sx,sy=RW_OU(p,n,K)  
ax2.set_title("K="+str(K)) 

for j in range(0,p):
    ax2.plot(sx[j], sy[j])
ax2.plot(0, 0, 'go')

n  = 500
p  = 2
K = 0.3
sx,sy=RW_OU(p,n,K)  
ax3.set_title("K="+str(K)) 

for j in range(0,p):
    ax3.plot(sx[j], sy[j])
ax3.plot(0, 0, 'go')
```
